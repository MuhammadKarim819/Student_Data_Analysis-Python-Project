# drop unnamed column
df=df.drop("Unnamed: 0", axis=1)
print(df.head())

# change weekly study hours column (Note 05-Oct is not exist. 5-10 is already have been written in this data so no change will occur we can skip this step)
df["WklyStudyHours"]= df["WklyStudyHours"].str.replace("05-Oct","5-10") 

```python
# Gender distribution
plt.figure(figsize= (5,5))
ax= sns.countplot(data= df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

# from the above chart I have analysed that:
# numbers of females are more than the numbers of male

```python
# Parent impact on scores
gb= df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
# Show the heatmap for these value
plt.figure(figsize= (5,5))
sns.heatmap(gb, annot = True)
plt.title("Parents Education impact on students Scores")
plt.show()
```

# From the above cahrt I Have analysed that the education of parents have higly impact on the the students education 

```python
# Parent Marital Status impact on scores
gb1= df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
# Show the heatmap for these value
plt.figure(figsize= (5,5))
sns.heatmap(gb1, annot = True)
plt.title("Parent Marital Status impact on scores")
plt.show()
```

# from the above charts I analysed that there is no/neglible imapct on studdents scores due to marital status

```python
# Distribution of Ethnic Groups
groupA = df.loc[(df["EthnicGroup"]=="group A")].count()
groupB = df.loc[(df["EthnicGroup"]=="group B")].count()
groupC = df.loc[(df["EthnicGroup"]=="group C")].count()
groupD = df.loc[(df["EthnicGroup"]=="group D")].count()
groupE = df.loc[(df["EthnicGroup"]=="group E")].count()

l=["group A", "group B", "group C", "group D", "group E"]
mlist=[groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]

plt.pie(mlist,labels=l, autopct= "%1.2f%%")
plt.title("Distributuion of EthnicGroup")
plt.show()
```
